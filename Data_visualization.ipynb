```python
# Scatter plot
plt.scatter(data['Age'], data['EstimatedSalary'])
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Age vs Estimated Salary')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='EstimatedSalary', data=data)
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Age vs Estimated Salary')
plt.show()
```

```python
# Histogram for EstimatedSalary
plt.hist(data['EstimatedSalary'], bins=20)
plt.xlabel('Estimated Salary')
plt.ylabel('Frequency')
plt.title('Estimated Salary Distribution')
plt.show()
```

```python
# Pie chart
# Calculate counts for different age groups
age_counts = data['Age'].value_counts()

# Creating plot
fig = plt.figure(figsize=(6, 6))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart: Age Distribution')

# Show plot
plt.show()
```
